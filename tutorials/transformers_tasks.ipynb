### ner (named entity recognition)

Note:
* the modelsheet;
* model parameters;
* token;

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple", model="DeepMount00/Italian_NER_XXL")
text = """Il commendatore Gianluigi Alberico De Laurentis-Ponti, con residenza legale in Corso Imperatrice 67,
          Torino, avente codice fiscale DLNGGL60B01L219P, è amministratore delegato della "De Laurentis Advanced
          Engineering Group S.p.A.",  che si trova in Piazza Affari 32, Milano (MI); con una partita IVA di 09876543210,
          la società è stata recentemente incaricata  di sviluppare una nuova linea di componenti aerospaziali
          per il progetto internazionale di esplorazione di Marte."""


pd.DataFrame(ner_tagger(text))
```

### question-answering

Note:
* default model for task;

```python
context = """
          Earth's rotation or Earth's spin is the rotation of planet Earth around its own axis,
          as well as changes in the orientation of the rotation axis in space. Earth rotates eastward,
          in prograde motion. As viewed from the northern polar star Polaris, Earth turns counterclockwise.
          """

reader = pipeline("question-answering")
reader(context=context, question="how rotate earth")
```

### summarization

Note:

* Model specific parameter

```python
text = """
       La Rivoluzione francese fu un periodo di sconvolgimento sociale, politico e culturale estremo, e
       prevalentemente violento, avvenuto in Francia tra il 1789 e il 1799, poi allargatosi in Europa con
       le guerre rivoluzionarie francesi e le guerre napoleoniche. In storiografia è lo spartiacque tra età
       moderna ed età contemporanea.[1] È detta anche Prima rivoluzione francese o Grande rivoluzione
       francese, per distinguerla dalla Rivoluzione di luglio del 1830 (Seconda rivoluzione francese),
       dai moti rivoluzionari del 1848 (Terza rivoluzione francese), che furono l'episodio locale del periodo
       di rivolte europee conosciuto come la Primavera dei popoli, e dalla Comune di Parigi del 1871.
       """
summarizer = pipeline("summarization", model='csebuetnlp/mT5_multilingual_XLSum', max_length=100)
summarizer(text)
```

### audio-classification

```python
from datasets import load_dataset, Audio

dataset = load_dataset("PolyAI/minds14", name="it-IT", split="train")
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))
sampling_rate = dataset.features["audio"].sampling_rate
audio_file = dataset[0]["audio"]["path"]
```

```python
from IPython.display import Audio, display

display(Audio(audio_file, autoplay=True))
```

### text to speech

Note:
* tensor offloading;
* dependencies and retrocompatibility;

```python
%%time

text="""[clears throat] This is a test ... and I just took a long pause."""

synth = pipeline("text-to-speech")
speech=synth(text)
speech
```

```python
%%time
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

synth = pipeline("text-to-speech", device=device)
speech=synth(text)
speech
```

```python
#import soundfile as sf
#sf.write("speech.wav", speech["audio"].T, samplerate=speech['sampling_rate'])

import scipy
scipy.io.wavfile.write("speech.wav", rate=speech["sampling_rate"], data=speech['audio'].T)


from google.colab import files
files.download('speech.wav')
```

### text to music

```python
synth = pipeline("text-to-audio", "facebook/musicgen-small", device=device)
text = "a chill song with influences from lofi, chillstep and downtempo"

music = synth(text, forward_params={"do_sample":True})
```

```python
music["audio"].shape
```

```python
import scipy
from google.colab import files
scipy.io.wavfile.write("music.wav", rate=music["sampling_rate"], data=music['audio'])
files.download("music.wav")
```
